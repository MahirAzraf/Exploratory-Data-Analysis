Data Cleaning and Preprocessing

```python
df.columns = df.columns.str.strip()

# Step 2: Remove missing or blank product names
df.dropna(subset=['product_name'], inplace=True)

# Step 3: Remove duplicate rows
df.drop_duplicates(inplace=True)

# Step 4: Strip white spaces and make product names lowercase (standardize for analysis)
df['product_name'] = df['product_name'].str.strip().str.lower()

# Step 5: Convert rating and rating_count to numeric (some might be stored as strings)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# Step 6: Drop rows where rating count is missing (optional, but helps in analysis)
df.dropna(subset=['rating_count'], inplace=True)

# Preview cleaned data
print("✅ Data cleaned successfully!")
df.head()
```

```python
print(df.isnull().sum())
```

```python
# Drop the row with missing rating
df.dropna(subset=['rating'], inplace=True)

# Check again
print("Null values after cleaning:")
print(df.isnull().sum())
```

```python
# Count how many times each product appears
product_counts = df['product_name'].value_counts()

# Show the top 10 most used products
product_counts.head(10)
```

```python
import matplotlib.pyplot as plt

# Plot the top 10 most used products
product_counts.head(10).plot(kind='bar', figsize=(10, 5), color='skyblue')
plt.title('Top 10 Most Used Products by Customers')
plt.xlabel('Product Name')
plt.ylabel('Number of Uses')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Sort products by rating in descending order
top_rated_products = df.sort_values(by='rating', ascending=False)

# Drop duplicates so we see only one entry per product
top_rated_unique = top_rated_products.drop_duplicates(subset='product_id')

# Show top 10 highest rated unique products
top_rated_unique[['product_name', 'rating', 'rating_count', 'discounted_price', 'category']].head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select top 10 highest rated unique products
top_10 = top_rated_unique[['product_name', 'rating']].head(10)

# Set figure size
plt.figure(figsize=(12, 6))

# Create a barplot
sns.barplot(x='rating', y='product_name', data=top_10, palette='viridis')

# Add titles and labels
plt.title('Top 10 Best-Rated Products', fontsize=16)
plt.xlabel('Rating')
plt.ylabel('Product Name')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Group by product and calculate average rating and total number of reviews
rating_summary = df.groupby('product_name').agg({
    'rating': 'mean',
    'rating_count': 'sum'
}).reset_index()

# Keep only products with at least 50 reviews (optional filter to avoid low-sample products)
popular_rated = rating_summary[rating_summary['rating_count'] >= 50]

# Sort by rating (you can also consider sorting by a combination if needed)
top_combined = popular_rated.sort_values(by='rating', ascending=False).head(10)

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(x='rating', y='product_name', data=top_combined, palette='coolwarm')

plt.title('Top 10 Highest Rated Products (With at least 50 Reviews)', fontsize=16)
plt.xlabel('Average Rating')
plt.ylabel('Product Name')

plt.tight_layout()
plt.show()
```
